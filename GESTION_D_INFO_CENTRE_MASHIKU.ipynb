### 3) À L’AIDE DE L’ALGORITHME 𝑘-MEANS, SEGMENTONS L'ENSEMBLE DE DONNEES EN DIX CLASSES

```python
# Imortation des bibliothèques
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Construction et application du modèle K-Means
model= KMeans(n_clusters=10)

# Adaptation du modèle de données
model.fit(data)

# La couleur des points dépend de leur étiquette (label)
plt.scatter(data.ID_individu,data.fixed_acidity, c=model.labels_, s=5)
plt.colorbar()

# Affichage du résultat
# Les centres de cluster ont une couleur bleu
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_ [:,1],c='blue',s=5)
print(plt.show())

# Affichages de toutes les composantes
print(model.cluster_centers_)
```

```python
# Affectons les données segmentées
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)

# Construction du modèle de classificateur de la regression logistique provenant de sklearn
#Grace à l'attribut solver, on fait appel à des méthodes de minim
classifier = LogisticRegression(solver='lbfgs',random_state=0)

# Transformation des données d'entrainement dans le classificateur afin qu'il puisse régler ses paramètres interne et être prêt pour les prédictions futures des données.
classifier.fit(X_train, Y_train) # Application de la mise à l'échelle sur les données d'apprentissage

# La sortie d'exécution
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True, intercept_scaling=1, max_iter=100, multi_class='warn', n_jobs=None, penalty='l2', random_state=0, solver='lbfgs', tol=0.0001, verbose=0, warm_start=False)

# Appel de la méthode de prédiction
predicted_y = classifier.predict(X_test)

print('L\'exactutide du modèle est de : {:.2f}'.format(classifier.score(X_test, Y_test)))
print('Soit en pourcentage (%): {:.2f}'.format(classifier.score(X_test, Y_test)*100))
print('')
MatriceDeConfusion=confusion_matrix(Y_test,predicted_y)
print('')
print('Voici la matrice de confusion :')
print(MatriceDeConfusion)
```
